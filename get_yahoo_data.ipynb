```python
import yfinance as yf
import pandas as pd

spy = yf.Ticker('SPY').funds_data
spy.description
spy.top_holdings
```

```python
df = pd.read_csv('data/spy_holdings_24.csv')
df.head()

```

```python
for i in range(len(df)):
    
    tick = df.iloc[i]['Ticker']
    fn = f'data/{tick}.csv'
    dfi = yf.download(tick)
    dfi2 = dfi.stack(level=1).rename_axis(['Date', 'Ticker']).reset_index(level=1)
    dfi2.to_csv(fn)    

```

```python
# check broken tickers

tick_count = []

for i in range(len(df)):
    
    tick = df.iloc[i]['Ticker']
    fn = f'data/{tick}.csv'
    dfi = pd.read_csv(fn)
    tick_count.append((tick,len(dfi)))

#print(tick_count)
bad_ticks = list(filter(lambda x: x[1] == 0, tick_count))
print(len(bad_ticks))
print(bad_ticks)


```

```python


for i in range(len(bad_ticks)):
    tick = bad_ticks[i][0]
    tick2 = tick.replace('.','-')
    fn = f'data/{tick}.csv'
    dfi = yf.download(tick2)
    dfi2 = dfi.stack(level=1).rename_axis(['Date', 'Ticker']).reset_index(level=1)
    dfi2.to_csv(fn)
    
```
